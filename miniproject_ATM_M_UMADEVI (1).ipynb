```python
from abc import ABC, abstractmethod# immporting ABC(Abstarct Base Class)and Abstract method
class BankAccount(ABC):#creating an abstract class
    @abstractmethod#using an abstract method annotation
    def transaction(self):#creating an abstract method
        pass
class Account:#creating a parent class
    def __init__(self, name, ac_no, bal, pin):#defining a constructor
        self.name = name
        self._account_number = ac_no
        self.__balance = bal
        self.__pin = pin
        self.history = []
        self.count = 0
    def get_bal(self):            #usinng get method toreturn private balance
        return self.__balance
    def set_bal(self, val):       #Using set method to update balance
        self.__balance = val

    def check_p(self, input_p):     #using check method to check the correct pin
        return self.__pin == input_p
class ATM(Account, BankAccount):      #creating a child class
    @classmethod
    def bank_info(cls):
        print("\n------------------------------")
        print("    WELCOME TO ABC BANK ATM   ")#print welcome message
        print("------------------------------")
    @staticmethod
    def check_amt(amt):               #to check amount is positive or not
        return amt > 0
    def transaction(self):            # implementation of abstract method
        print("UMA account total balanc...")
        self.count +=1
    def check_balance(self):          #checking the balance
        print("Balance: â‚¹", self.get_bal())
        self.history.append("Checked Balance")
    def deposit(self, amt):               #using the deposit method
        if self.check_amt(amt):
            new_val = self.get_bal() + amt
            self.set_bal(new_val)                 #adding amount to previous balance
            self.history.append(f"Added: {amt}")
            print("Deposit Done UMA!")
        else:
            print("Invalid amount!")
    def withdraw(self, amt):                      #using with draw method
        if amt > 5000:
            print("Max limit is 5000!")
            return

        if self.check_amt(amt) and amt <= self.get_bal(): #checking comdition
            new_val = self.get_bal() - amt
            self.set_bal(new_val)
            self.history.append(f"Took: {amt}")   #adding amount
            print("Withdrawal Done! RAHUL")
        else:
            print("No enough money or invalid amount!")

    def view_info(self):                      #viewing account information
        print("\n--- Account Info ---")
        print("User:", self.name)
        print("A/C No:", self._account_number)
        print("Total Actions:", self.count)
        print("History:", self.history)
def main():                               #program execution strats here
    obj = ATM("Rahul", 12345, 1000, "1122")
    p = input("Enter PIN for UMA account: ")
    if not obj.check_p(p):
        print("Wrong PIN!")
        return

    while True:
        obj.bank_info()
        print("1. Check Balance")
        print("2. Deposit Money")
        print("3. Withdraw Money")
        print("4. Account Details")
        print("5. Exit")
        choice = input("\nSelect (1-5): ")
        #using else if condition
        if choice == '1':
            obj.transaction()
            obj.check_balance()
        elif choice == '2':
            amt = float(input(" Enter Amount to deposit UMA: "))
            obj.deposit(amt)
        elif choice == '3':
            amt = float(input("Enter Amount to withdraw UMA: "))
            obj.transaction()
            obj.withdraw(amt)
        elif choice == '4':
            obj.view_info()
        elif choice == '5':
            #
            ok = input("Are you sure? (y/n): ")
            if ok == 'y':
                print("Bye Bye!")
                break # [5]
        else:
            print("Try again!")

if __name__ == "__main__":    #to control the python program
    main()
```
